### Статья N1: A robust deep learning approach for glasses detection in non-standard facial images
Link: https://ietresearch.onlinelibrary.wiley.com/doi/full/10.1049/bme2.12004


<div style="text-align: justify">
В предшествующих исследованиях обнаружение очков всегда проводилось в экспериментах либо с использованием самостоятельно сгенерированных фичей (Табл.1) , либо с использованием фичей на основе глубокого обучения. Тем не менее, в обоих случаях для требовались очень стандартные изображения лиц/глаз. Оба рабочих метода работают достаточно хорошо, но качество работы алгоритма очень сильно зависело от качества изображений лиц и извлеченных признаков, где небольшое смещение и/или поворот лица отрицательно сказывались на результатах. </div>
<br>
<div style="text-align: justify">
Кроме того, качество на реальных (нестандартных) изображениях становится еще хуже. В этой статье авторы предлагают надежный подход к глубокому обучению для обнаружения очков на селфи-фотографиях с полной или частичной фронтальной частью тела, нестандартных изображений, снятых в реальных неконтролируемых условиях, без использования какие-либо лицевых ориентиров на основе CNN и transfered learning.

#### Рис.2 Предложенная архитектура 

<div style="text-align: justify">
В отличие от предыдущих публикаций Табл.1, В этой статье используется реалистичный и сложный датасет: 46K селфи, с несинтезированными очками и сложными фронтальными лицами с полным / частичным телом (высокая изменчивость - фото сделанные обычными пользователями для себя в любое время и в любом месте). Предложенная модель CNN достигла accuracy 96%. Чтобы достичь такого результата, предлагаемая модель была построена на основе transfer learing с 1,2-миллионного набора данных ImageNet, что позволяет использовать абстракцию необработанных данных, которая расширяет анализ до ранее незнакомых данных.</div>
<br>
<div style="text-align: justify">
Однако даже при таком трансферном обучении сети приходилось иметь дополнительные слои и проходить полные эпохи (полный цикл обучения на всех обучающих данных), растянувшиеся почти на две недели. После такого обширного обучения сеть изучила множество различных компонентов изображения, то есть края и цветные пятна, которые важны для обнаружения наличия очков на входном изображении.</div>
<br>
<div style="text-align: justify">
Использование трансферного обучения в сочетании с новыми feature map компенсирует спорадический характер селфи-изображений. Это позволяет сети достичь более высокой скорости обнаружения очков.</div>


### Статья N2: Shallow Convolutional Neural Network for Eyeglasses Detection in Facial Images Arwa
Link: https://www.researchgate.net/publication/274249930_Glasses_detection_on_real_images_based_on_robust_alignment

<div style="text-align: justify">
Недавние исследования показывают, что переменные извлеченные на основе применения CNN дают убедительное качество работы алгоритма, что в то же время указывает на необходимость дополнительного исследования разных архитектур и паттернов проектирования и их взаимосвязь с эффективностью модели.</div>
<br>
<div style="text-align: justify">
Основной вклад этой статьи состоит в рассмотрении двух важные аспектов CNN: (1) требования к размеру обучающего набора данных и (2) глубине сетевой архитектуры.  В частности авторы инициализируют параметры обучения неглубокой CNN с помощью параметров глубокой CNN, которая точно настроена на небольшом наборе данных. Затем глубина нейронной сети уменьшается путем удаление некоторых сверточных слоев после тестирования на валидационном датасете. По результатам проделанной работы авторы получили новую неглубокую CNN с высокой accuracy -  Shallow-GlassNet. Помимо достижения высокой точности достигается не только высокая точность, модель отличает высокая скорость обнаружения очков.  Тестирования модели было проведено на двух больших базах данных изображений лиц, LFW and Celeb Faces. Результативность предлагаемой архитектутры достигает  99,73%.</div>
<br>
<div style="text-align: justify">
Вдохновленные GoogleNet архитектурой, предложенная модель Shallow-GlassesNet, as (Рис.1) состоит из 6 слоев (3 конволюционных, два макс пуллинга и fully-connected слой). Предложенный пайплайн состоит из трех частей : детекция лица, экстракция фичей c помощью Shallow Glasses Net, и классификация / детекция с помощью SVM.</div>


#### Рис.1 Предложенная архитектура 



<div style="text-align: justify"> 
Авторы использовали предобученную GoogleNet для дообучения на изображениях лиц с очками и без. Далее обученные веса GoogleNet копировались в соответствующие уровни Shallow-GlassesNet которая использовалась как фича экстрактор.
Линейный SVM обучался на извлеченных фичах для детекции очков. Предложенная Shallow-GlassesNet архитектура снизила время обнаружения очков на почти на два порядка с сохранением высокой accuracy. Таким образом ключевым вкладом данной работы является уменьшение времени детекции с сохранением высокого качества работы алгоритма (Табл.2).</div>


### Статья N3: Glasses detection on real images based on robust alignment
Link: https://ietresearch.onlinelibrary.wiley.com/doi/full/10.1049/bme2.12004
https://dl.acm.org/doi/abs/10.1007/s00138-015-0674-1

<div style="text-align: justify"> 
Автоматическое определение очков на реальных изображениях лица является сложной задачей из-за различных характеристик внешности. В этой работе предлагается использовать два метода для улучшения решения данной задачи определния наличия очков: Robust Local Binary Pattern и robust alignment. Во-первых, изображения предварительно обрабатываются и нормализуются, чтобы проблему масштаба и поворотов.
    
<br>Во-вторых, определяется область очков учитывая, что носовая часть очков обычно располагается на одном уровне с центром глаз как по высоте, так и по ширине. В-третьих, Robust Local Binary Pattern применяется для описания области глаз и наконец, SVM используется для классификации характеристик LBP (local binary pattern) Рис.3. Данный алгоритм из-за его надежности и скорости  можно использовать для задачи удаления очков Предложенный алгоритм был протестирован на датасетах Labeled Faces (Wild database) с recognition rate =  98.65%.</div>


#### Рис.3 LBP Operator 

<div style="text-align: justify"> 
Авторы продемонстрировали влияние расширения, выравнивания, нормализации изображения на качество работы LBP.  На LFW датасете качество составило  98.65%. Данный алгоритм также был протестирован на FERET (99.89% recognition rate). Эксперименты показали что предложенная модель устойчива к различным условиям освещения, позам людей, окклюзией (что часто встечается у солнцезащитных очков). Даже когда основные черты лица некорректно извлекаются, алгорим надежно извлекает регион расположения очков (Рис.4).
</div>
